# **Average Comparison (One-Sample t-test)**
Problem: A food delivery company claims that the average delivery time is 30 minutes.
You suspect it is actually higher. Collect a sample of delivery times. Test whether the true mean delivery time is greater than 30 minutes.

## **Defining Hypothesis :**
Null Hypothesis : mean = 30
Alternate hypotheis : mean>30

## **Sample data**

data = [32, 35, 30, 28, 34, 33, 31, 29, 36, 38] <br>
n= 10 <br>
mean =  (32+35+30+28+34+33+31+29+36+38) / 10 = 32.6<br>
Standard deviation ~ 3.2 minutes

## **T-test statistics**

t= (sample mean - null mean)/std deviation*sqrt(n) ~ 2.7 <br>
Significance level = 5% = 0.05 <br>
degrees of freedom = n-1 = 10-1 = 9

For degree = 9 tcritical = 1.833

Here t (2.57) > tcritical (1.833) hence reject hypthesis

## **Python overview**

```python
import scipy.stats as stats

data = [32, 35, 30, 28, 34, 33, 31, 29, 36, 38]

t_test, p_value = stats.ttest_1samp(data,30)

#since it is one sample t test,
p_value/=2
print(t_test,p_value)
```

Here p_value < 0.05, which means we reject the hypothesis
