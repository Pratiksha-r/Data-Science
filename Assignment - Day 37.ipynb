```python
# From the data provided on Hollywood movies:
# 1.	Find the highest rated movie in the “Quest” story type.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data=pd.read_csv("/content/HollywoodMovies.csv")
data_score=data[(data['Story']=='Quest') | (data['Story']=='quest')][['Movie','AudienceScore','Story']]
data_score[data_score['AudienceScore']==data_score['AudienceScore'].max()][['Movie','AudienceScore']]
```

```python
# 2.	Find the genre in which there has been the greatest number of movie releases
# Genre and number of movies in each Genre

genre_movie=data.groupby('Genre')['Movie'].count()
genre_movie[genre_movie.values==genre_movie.values.max()]
```

```python
# 3.	Print the names of the top five movies with the costliest budgets.
data_sort=data.sort_values('Budget', ascending=False)
data_sort.head(5)['Movie']
```

```python
# 4.	Is there any correspondence between the critics’ evaluation of a movie and its acceptance by the public? Find out, by plotting the net profitability of a
# movie against the ratings it receives on Rotten Tomatoes.

plt.scatter(data['Profitability'],data['RottenTomatoes'])
plt.xlabel('Profitability')
plt.ylabel('Rating')
plt.show()

```

```python
# 6. Perform Operations on Files

# 6.1: From the raw data below create a Pandas Series

# 'Amit', 'Bob', 'Kate', 'A', 'b', np.nan, 'Car', 'dog', 'cat'
# a)	Print all elements in lower case
data=pd.Series(['Amit', 'Bob', 'Kate', 'A', 'b', np.nan, 'Car', 'dog', 'cat'])
data

# b)	Print all the elements in upper case
print(data.str.upper())


# c)	Print the length of all the elements
print(data.str.len())
```

```python
# 6.2: From the raw data below create a Pandas Series
# ' Atul', 'John ', ' jack ', 'Sam'
# a.	Print all elements after stripping spaces from the left and right
data=pd.Series([' Atul', 'John ', ' jack ', 'Sam'])
print(data.str.strip())
print(data.str.strip().str.len())

# b.	Print all the elements after removing spaces from the left only
print(data.str.lstrip())
print(data.str.lstrip().str.len())

# c.	Print all the elements after removing spaces from the right only
print(data.str.rstrip())
print(data.str.rstrip().str.len())
```

```python
# 6.3: - Create a series from the raw data below
# 'India_is_big', 'Population_is_huge', np.nan, 'Has_diverse_culture'

# a)	split the individual strings wherever ‘_’ comes and create a list out of it.
arr=pd.Series(['India_is_big', 'Population_is_huge', np.nan, 'Has_diverse_culture'])
print(arr.str.split("_"))

# b)	Access the 1st Element individual element of a list
print(arr.str.split("_").str[0])

# c)	Expand the elements so that all individual elements get splitted by ‘_’ and insted of list returns individual elements
print(arr.str.split("_",expand=True))

```

```python
# 6.4 Create a series

# A', 'B', 'C', 'AabX', 'BacX','', np.nan, 'CABA', 'dog', 'cat'
#  and replace either a (a is second character in a line, then replace first and second character )  or dog with XX-XX'

data=pd.Series(['A', 'B', 'C', 'AabX', 'BacX','', np.nan, 'CABA', 'dog', 'cat'])
print(data.str.replace("^.a|dog","XX-XX",case=False))
```

```python
# 6.5: Create a series and remove dollar from the numeric values '12', '-$10', '$10,000'
data=pd.Series(['12', '-$10', '$10,000'])
print(data.str.replace("$",""))
```

```python
# 6.6: Create a series and reverse all lower-case words

# 'india 1998', 'big country', np.nan
data=pd.Series(['india 1998', 'big country', np.nan])
replacement = lambda x: x.group(0)[::-1]

print(data.str.replace('[a-z]+', replacement, regex=True))
```

```python
# 6.7: Create pandas series and print true if value is alphanumeric
# in series or false if value is not alpha numeric in series.
# '1', '2', '1a', '2b', '2003c'

data=pd.Series(['1', '2', '1a', '2b', '2003c'])
print(data.str.isalnum())
```

```python
# 6.8: Create pandas series and print true if value is containing ‘A’

# '1', '2', '1a', '2b', 'America', 'VietnAm','vietnam', '2003c'

data=pd.Series(['1', '2', '1a', '2b', 'America', 'VietnAm','vietnam', '2003c'])
print(data.str.contains('A', case=False))
```

```python
# 6.9: Create pandas series and print in three columns value 0 or 1 is a or b or c exists in values

# 'a', 'a|b', np.nan, 'a|c'

data=pd.Series(['a', 'a|b', np.nan, 'a|c'])
regex='[a|b|c]'
print(data.str.get_dummies(sep='|'))
```
