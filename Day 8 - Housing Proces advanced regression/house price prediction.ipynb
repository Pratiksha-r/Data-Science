```python
train_data.drop(columns=['Id','Alley','Fence','MiscFeature'],inplace=True)
train_data['LotFrontage'] = train_data['LotFrontage'].fillna(train_data['LotFrontage'].mean())

train_data['GarageType']=train_data['GarageType'].fillna('NA')
train_data['GarageYrBlt']=train_data['GarageYrBlt'].fillna(int(0))
train_data['GarageFinish']=train_data['GarageFinish'].fillna('NA')
train_data['GarageQual']=train_data['GarageQual'].fillna('NA')
train_data['GarageCond']=train_data['GarageCond'].fillna('NA')


train_data['BsmtQual']=train_data['BsmtQual'].fillna('NA')
train_data['BsmtCond']=train_data['BsmtCond'].fillna('NA')
train_data['BsmtExposure']=train_data['BsmtExposure'].fillna('NA')
train_data['BsmtFinType1']=train_data['BsmtFinType1'].fillna('NA')
train_data['BsmtFinType2']=train_data['BsmtFinType2'].fillna('NA')


train_data['PoolQC']=train_data['PoolQC'].fillna('NA')
train_data['FireplaceQu']=train_data['FireplaceQu'].fillna('NA')


train_data['MasVnrType']=train_data['MasVnrType'].fillna(train_data['MasVnrType'].mode()[0])
train_data['MasVnrArea']=train_data['MasVnrArea'].fillna(train_data['MasVnrArea'].mean())

train_data['Electrical']=train_data['Electrical'].fillna(train_data['Electrical'].mode()[0])
```

```python
print(train_data.isnull().sum()[train_data.isnull().sum()>0])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

s=(train_data.dtypes == 'object')
object_cols=list(s[s].index)

lr=OneHotEncoder(sparse_output=False)
encoded_data=pd.DataFrame(lr.fit_transform(train_data[object_cols]))
encoded_data.index=train_data.index
encoded_data.columns=lr.get_feature_names_out()
train_data=train_data.drop(columns=object_cols, axis=1)
train_data=pd.concat([train_data,encoded_data],axis=1)
```

```python
train_data.head()
```

```python
x=train_data.drop(columns=['SalePrice'])
y=train_data['SalePrice']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor

RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
```

```python
y_pred=RFR.predict(x_test)

from sklearn.metrics import r2_score

print('Accuracy :', r2_score(y_pred,y_test))
print('Mean absolute square : ',mean_absolute_error(y_pred,y_test))
```

```python
from sklearn.linear_model import SGDRegressor
SGD=SGDRegressor()
SGD.fit(x_train,y_train)
```

```python
y_pred=SGD.predict(x_test)

from sklearn.metrics import r2_score

print('Accuracy :', r2_score(y_pred,y_test))
print('Mean absolute square : ',mean_absolute_error(y_pred,y_test))
```

```python
X_train = np.array(x_train)
X_test = np.array(x_test)
```

```python
import warnings
warnings.filterwarnings('ignore')

from xgboost import XGBRegressor
xgb=XGBRegressor(learning_rate=0.03,n_estimators=200,objective='reg:squarederror')

xgb.fit(X_train,y_train)
y_pred=xgb.predict(X_test)

from sklearn.metrics import r2_score

print('Accuracy :', r2_score(y_pred,y_test))
print('Mean absolute square : ',mean_absolute_error(y_pred,y_test))
```

```python
plt.figure()
plt.title("Comparison of Sale Price of Predicted and Actual Values")
plt.scatter(y_test,RFR.predict(x_test), label='Random Forest')
plt.scatter(y_test,xgb.predict(X_test), label='XGB')
plt.legend()
```

```python
from sklearn.model_selection import GridSearchCV
model=XGBRegressor()

n_estimators=[100,200,300]
learning_rate=[0.03,0.1,0.3]
objectives= ['reg:squarederror']

hyperparameter_grid = {
    'n_estimators' : n_estimators,
    'learning_rate' : learning_rate,
    'objective' : objectives
}

grid_cv=GridSearchCV(estimator=model,
                     param_grid=hyperparameter_grid,
                     scoring='neg_mean_absolute_error',
                     return_train_score=True)

grid_cv.fit(X_train,y_train)
```

```python
print(grid_cv.best_score_)
print(grid_cv.best_estimator_)
```
