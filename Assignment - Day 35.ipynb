```python
# 1.	You are given a dataset, containing the number of hurricanes (Hurricanes.csv) occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data,
# taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data=pd.read_csv("/content/Hurricanes.csv")
data.head()

plt.figure(figsize=(10,8))
plt.bar(data['Year'],data['Hurricanes'])
plt.show()
```

```python
from numpy import rot90
# 3.1 Load the Data set sample-salesv2.csv , then display top 5 records
sample_sales=pd.read_csv("/content/sample-salesv2.csv")
print(sample_sales.head(5))

# 3.2 Use the Describe statement and describe the column – Unit Price
sample_sales['unit price'].describe()

# 3.3 Extract the columns / create  a data frame customers , with columns Name, Net_price . Do the group by based on column name
customers = sample_sales[['name','net_price']]
customer_group = customers.groupby('name')
print(customer_group.size())


# 3.4 – Create bar graph, customer by Sales ( x- axis – Customer , y-axis – Sales)
sum_value=customer_group.sum()
sum_value.plot(kind='bar')
```

```python
from matplotlib import axes
from matplotlib.lines import lineStyles
# 4.	Let the x axis data points and y axis data points are
X = [1,2,3,4]
y = [20, 21, 20.5, 20.8]
fig, axes = plt.subplots(1,2, figsize=(10,4))

# 4.1: Draw a Simple plot
axes[0].plot(X,y)


# 4.2: Configure the line and markers in simple plot
axes[1].plot(X,y,linestyle='dashed',marker='o')

# 4.3: configure the axes
plt.xticks(rotation=90)
plt.xticks([1,2,3,4,5,6])
plt.xlim(0.5,4.5)

# 4.4: Give title of Graph & labels of x axis and y axis
plt.title("Title of the graph")
plt.xlabel("Title of x axis")
plt.ylabel("Title of y label")

# 4.5: Give error bar if y_error = [0.12, 0.13, 0.2, 0.1]
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]
y_error = [0.12, 0.13, 0.2, 0.1]
fig, axes = plt.subplots(1,2,figsize=(10,4))
axes[0].plot(x,y)
axes[1].errorbar(x,y,y_error)



# 4.6: define width, height as figsize=(4,5) DPI and adjust plot dpi=100
fig = plt.figure(figsize=(4, 5), dpi=100)
y_error = [0.12, 0.13, 0.2, 0.1]
plt.plot(x, y, linestyle="dashed", marker="o", color="green")
plt.errorbar(x, y, yerr=y_error, linestyle="None", marker="None", color="blue")
plt.xlim(0.5,4.5)
plt.xticks([1,2,3,4])
plt.ylim(19.8,21.2)
plt.yticks([20, 21, 20.5, 20.8])

plt.xlabel("this is X")
plt.ylabel("this is Y")



# 4.7: Give a font size of 14
fig = plt.figure(figsize=(4, 5), dpi=100)
plt.plot(x, y, linestyle="dashed", marker="o", color="green")
plt.xlim(0.5,4.5)
plt.xticks([1,2,3,4])
plt.ylim(19.8,21.2)
plt.yticks([20, 21, 20.5, 20.8])

plt.rc('font',size=14)
```

```python
# 4.8: Draw a scatter graph of any 50 random values of x and y axis
import numpy as np
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors = np.random.rand(N)

area = np.pi * (10 * np.random.rand(N))**2
plt.scatter(x,y,c=colors, alpha=0.5,s=area)
plt.show()
```

```python
# 4.9: Create a dataframe from following data
# 'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
# 'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
# 'female': [0, 1, 1, 0, 1],
# 'age': [42, 52, 36, 24, 73],
# 'preTestScore': [4, 24, 31, 2, 3],
# 'postTestScore': [25, 94, 57, 62, 70]
# Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age

raw={'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]}
data=pd.DataFrame(raw)
plt.scatter(data['preTestScore'],data['postTestScore'], s=data['age']*2)
```
